```python
plt.pie(
    nbSexes,
    labels=['Homme' if x == 'M' else 'Femme' for x in nbSexes.index],
    autopct=lambda z: str(round(z, 2)) + '%',
    startangle=90,
    explode=[0.01,0.01],
    textprops={'fontsize': 15})
plt.title('Proportion Homme/Femme chez les Tenracs')

plt.show()
```
